## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Creating an object of class figure
fig = plt.figure()
# This is like an imaginary blank canvas

# adding a set of axes to the figure
ax = fig.add_axes([0,0,1,1])

# plotting the data x, y on the axes
ax.plot(x, y)

# adding label to the x axis
ax.set_xlabel('x')

# adding label to the y axis
ax.set_ylabel('y')

# adding title to the figure
ax.set_title("title")

# setting the limits on x and y axes
ax.set_xlim(left=0, right=100)
ax.set_ylim(bottom=0, top=200)

# setting the tick labels on x and y axes
ax.set_xticks(np.arange(start=0, stop=101, step=20))
ax.set_yticks(np.arange(start=0, stop=201, step=50))
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Creating an object of class figure
fig = plt.figure()
# This is like an imaginary blank canvas

# adding a first set of axes to the canvas
ax1 = fig.add_axes([0, 0, 1, 1])

# adding a second set of axes to the canvas
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# plotting data x, y on both the axes
ax1.plot(x, y, color='red')
ax2.plot(x, y, color='red')

# setting the limits for both set of az
ax1.set_xlim(left=0, right=100)
ax1.set_ylim(bottom=0, top=200)
ax2.set_xlim(left=0, right=100)
ax2.set_ylim(bottom=0, top=200)

# setting the tick labels on the smaller set of axes
ax1.set_xticks(np.arange(start=0, stop=101, step=20))
ax1.set_yticks(np.arange(start=0, stop=201, step=50))

ax2.set_xticks(np.arange(start=0, stop=101, step=20))
ax2.set_yticks(np.arange(start=0, stop=201, step=50))

# setting the labels for both set of axes
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
# creating an object of class figure
fig = plt.figure()

# adding a set of axes to the figure
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# plotting data on the axis ax1
ax1.plot(x,z)

# plotting data on the axis ax2
ax2.plot(x,y)

# adding title to ax2
ax2.set_title('zoom')

# adding labels to ax1
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

# adding labels to ax2
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

# setting limit for ax1
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

# setting limits for ax2
ax2.set_xlim([20,22])
ax2.set_ylim([30, 50])
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Creating a subplot with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2)

# plotting data (x,y) on the first axis
axes[0].plot(x, y, color='blue', ls='--', lw=3)

# plotting data (x,z) on the second axis
axes[1].plot(x, z, color='red', lw='3')

# setting the limit on axes[0] i.e. graph in column 1
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

# setting the limit on axes[1] i.e. graph in column 2
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])

# setting the tick labels on axes[0] i.e. graph in column 1
axes[0].set_xticks(np.arange(0, 101, 20))
axes[0].set_yticks(np.arange(0, 201, 50))

# setting the tick labels on axes[1] i.e. graph in column 2
axes[1].set_xticks(np.arange(0, 101, 20))
axes[1].set_yticks(np.arange(0, 10001, 2000))

# spacing the two graphs appropriately
plt.tight_layout()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# Creating a subplot with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(11, 3))

# plotting data (x,y) on the first axis
axes[0].plot(x, y, color='blue', lw=3)

# plotting data (x,z) on the second axis
axes[1].plot(x, z, color='red', ls='--', lw='3')

# setting the limit on axes[0] i.e. graph in column 1
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

# setting the limit on axes[1] i.e. graph in column 2
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])

# setting the tick labels on axes[0] i.e. graph in column 1
axes[0].set_xticks(np.arange(0, 101, 20))
axes[0].set_yticks(np.arange(0, 201, 50))

# setting the tick labels on axes[1] i.e. graph in column 2
axes[1].set_xticks(np.arange(0, 101, 20))
axes[1].set_yticks(np.arange(0, 10001, 2000))

# setting the label on axes[0] i.e. graph in column 1
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# setting the label on axes[1] i.e. graph in column 2
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

# spacing the two graphs appropriately
plt.tight_layout()
```
